- Formulae: data < min and data > max

data[~ data < min or data > max]

- IQR - InterQuantileRange - Q3-Q1 -> 0.25 - 0.75
- Q1 = 0.25
- Q3 = 0.75
- min = Q1 - 1.5 * IQR = Q1 - 1.5 * (Q3 - Q1)
- max = Q3 + 1.5 * IQR = Q1 - 1.5 * (Q3 - Q1)

```python
Q1 = df5['Time in Session (minutes)'].quantile(0.25)
Q3 = df5['Time in Session (minutes)'].quantile(0.75)

IQR = Q3 - Q1

mn = Q1 - 1.5 * IQR
mx = Q3 + 1.5 * IQR
```

```python
x = df5['Time in Session (minutes)'][~ ((df5['Time in Session (minutes)'] > mx) | (df5['Time in Session (minutes)'] < mn))]
```

```python
plt.boxplot(x)
```

```python
(df5['Time in Session (minutes)'] > mx)
```

```python
df5['Time in Session (minutes)'][~(df5['Time in Session (minutes)'] < mn)]
```
