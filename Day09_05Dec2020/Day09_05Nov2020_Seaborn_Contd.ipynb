## Iris Flower DataSet

Toy Dataset: Iris Dataset: [https://en.wikipedia.org/wiki/Iris_flower_data_set]

- A simple dataset to learn the basics.
- 3 flowers of Iris species. [see images on wikipedia link above]
- 1936 by Ronald Fisher.
- Petal and Sepal: http://terpconnect.umd.edu/~petersd/666/html/iris_with_labels.jpg # * Objective: Classify a new flower as belonging to one of the 3 classes given the 4 features.
- Importance of domain knowledge.
- Why use petal and sepal dimensions as features?
- Why do we not use 'color' as a feature?

The four features of these 3 types of flowers
(Iris setosa, Iris versicolor, Iris virginica) are:

1. Sepal Length
2. Sepal Width
3. Petal Length
4. Petal Width

We will classify the given flower by using this Four Features

```python
sns.catplot(x = 'species', y = 'petal_width', data = iris, kind = 'swarm')
```

```python
sns.catplot(y = 'species', x = 'petal_width', data = iris, kind = 'swarm')
```

```python
sns.catplot(x = 'species', y = 'petal_width', data = iris, kind = 'box')
```

```python
sns.violinplot(x = 'species', y = 'petal_width', data = iris)
```

```python
setosa = iris[iris['species'] == 'setosa']['petal_width']

sns.kdeplot(x = setosa , data = iris)
```

```python
sns.violinplot(x = 'species', y = 'petal_length', data = iris)
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'reg')
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'hex')
```

```python
sns.pairplot(data = iris, hue = 'species')
```

```python
sns.pairplot(data = iris, hue = 'species', markers = ['*', '>', '^'], palette = sns.color_palette('rainbow', n_colors = 3))
```
